```python
    
def get_picture_names(filename, kind, every, movie_type, N_max):
    """Get picture names from movie."""
    picture_names = list()
    picturepath = "Pictures/"
    for t in np.arange(0, N_max, every):
        suffix = "-" + movie_type + f"-{t}-" + kind
        picturename = picturepath + filename + suffix + ".png"
        picture_names.append(picturename)
    return picture_names


```
